You need the pDeep data given as parquet files, but they are probably given as MGF files. Therefore, run `mmproteo --read-ions --thread-count 0 mgf2parquet` to convert the MGF files into table-like parquet files.

```python
import pandas as pd
from mmproteo.utils import log
from mmproteo.utils.utils import ensure_dir_exists
from mmproteo.utils.formats import mz
import glob
```

```python
base_path = '/scratch/mirko.krause/pdeep'
parquet_wildcard_path = base_path + '/mgf/*.parquet'
training_columns_path = base_path + "/training_columns"
merged_training_columns_path = training_columns_path + "/merged"
training_files_wildcard_path = training_columns_path + "/*.parquet"
```

```python
logger = log.DummyLogger(verbose=True)
```

```python
ensure_dir_exists(training_columns_path)
ensure_dir_exists(merged_training_columns_path)
```

```python
parquet_paths = glob.glob(parquet_wildcard_path)
parquet_paths
```

```python
df = pd.read_parquet(parquet_paths[0])
df.head(1).T
```

```python
del df
```

```python
charge_postfix_characters = "|" + "".join(str(i) for i in range(10))
charge_postfix_characters
```

```python
def post_processor(df: pd.DataFrame) -> pd.DataFrame:
    df[mz.FilteringProcessor.default_peptide_sequence_column_name] = \
        df['params__title'].str.rstrip(charge_postfix_characters).apply(list)
    return df
```

```python
filtering_results = mz.filter_files(input_file_paths=parquet_paths,
                            output_path=training_columns_path,
                            fdr=None,
                            skip_existing=False,
                            is_decoy_column_name=None,
                            fdr_column_name=None,
                            post_processor=post_processor,
                            thread_count=0,
                            logger=logger)
filtering_results[:1]
```

```python
pd.DataFrame(filtering_results)
```
